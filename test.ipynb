```python
import cv2

# Function to count the number of frames in a video
def count_frames(video_path):
    # Open the video file
    video = cv2.VideoCapture(video_path)
    
    if not video.isOpened():
        print("Error: Could not open video.")
        return
    
    # Get the total frame count
    frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Release the video object
    video.release()
    
    print(f"Number of frames: {frame_count}")

# Example usage
video_path = "/mnt/data/anmol/Lightpipe/output/dataset_v2/0001_color.mp4"  # Replace with the actual path to your video
count_frames(video_path)
```

```python
import matplotlib.pyplot as plt

# Open the video file
video = cv2.VideoCapture(video_path)

if not video.isOpened():
    print("Error: Could not open video.")
else:
    # Get the total frame count
    frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Set the video to the last frame
    video.set(cv2.CAP_PROP_POS_FRAMES, frame_count - 1)
    
    # Read the last frame
    ret, frame = video.read()
    if ret:
        # Convert the frame from BGR to RGB for matplotlib
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Display the frame
        plt.imshow(frame_rgb)
        plt.axis('off')  # Hide axes
        plt.title("Last Frame")
        plt.show()
    else:
        print("Error: Could not read the last frame.")
    
    # Release the video object
    video.release()
```
